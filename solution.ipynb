# Предобработка

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline


# делим данные на обучающую и тестовую выборки
X, y = data.iloc[:, :-1], data.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y, random_state=31, shuffle=True)
```

```python
# определяем некатегориальные фичи
numeric_cols = [column for column in data.columns if column not in object_cols and column!=data.columns[-1]]

# imputer для вставки пропущенных значений, среднее для некатегориальных и мода для категорильных
imputer = ColumnTransformer([
    ('numeric_imputer', SimpleImputer(strategy='mean'), numeric_cols),
    ('categorical_imputer', SimpleImputer(strategy='most_frequent'), object_cols)
    ],
)

# scaler для некатегориальных переменных и encoder для категориальных, используя onehot encoder
scaler_encoder = ColumnTransformer([
    ('numeric_scaler', StandardScaler(), list(range(len(numeric_cols)))),
    ('categorical_encoder', OneHotEncoder(sparse_output=False), list(range(len(numeric_cols), len(numeric_cols) + len(object_cols))))
    ],
)

# делаем вывод трансформеров в виде pandas.DataFrame
imputer.set_output(transform='pandas')
scaler_encoder.set_output(transform='pandas')

# создаем pipeline для предобработки данных
transformer = Pipeline([
    ('imputing', imputer),
    ('scaling_encoding', scaler_encoder),]
)
transformer.set_output(transform='pandas')
```

```python
# преобразуем данные
X_train_transformed = transformer.fit_transform(X_train)
X_test_transformed = transformer.transform(X_test)
```

```python
X_train_transformed.shape
```

# Обучение модели

Наша проблема стоит в бинарной классификации, первая идея - попробовать логистическую регрессию.   
Так как таргет несбалансирован есть смысл применить технику undersampling или/KFold кроссвалидацию, а также указать веса классов для наказания неверных предсказаний класса-меньшинства.  
Наиболее подходящая технология XGBoost в силу своей быстроты и эффективности, но я попробую и модели из sklearn: LogisticRegression, RandomForestClassifier.  
Выбор метрики очень важен, так recall очень важен если нужно поймать как можно больше положительных классов правильно, roc-auc score же определяет насколько модель умеет разделять классы.  
Так как основное напрявление вакансии fraud detection, то важно чтобы модель ловила как можно больше положительных классов(True Positive) и при этом как можно меньше помечала отрицательные классы как положительные(False Positive).  
Поэтому основная метрика - это roc-auc score.

```python
# пробуем логистическую регрессию, перебираем параметры gridsearh
lr_params = {
    'C': [0.5, 1, 2, 5],
    'class_weight': [{0:0.013, 1:1}, {0:1, 1:1}]
}
grid_lr = GridSearchCV(LogisticRegression(), param_grid=lr_params, verbose=2, n_jobs=-1, scoring='roc_auc')
grid_lr.fit(X_train_transformed, y_train)
```

```python
# ищем порог, который максимизирует roc-auc score
best_score, best_treshold = 0, np.nan
for t in np.linspace(0.1, 0.9, 20):
    y_pred = grid_lr.predict_proba(X_test_transformed)
    y_pred = y_pred[:, 1] > t
    if roc_auc_score(y_test, y_pred) > best_score:
        best_score, best_treshold = roc_auc_score(y_test, y_pred), t
    
y_pred = grid_lr.predict_proba(X_test_transformed)
y_pred = y_pred[:, 1] > best_treshold
print('Best treshold: ', best_treshold)
print(classification_report(y_test, y_pred))
print('Best roc-auc score: ', roc_auc_score(y_test, y_pred))
```

Повторяем процедуру для случайного леса

```python
rf_params = {
    'n_estimators': [100, 200, 300, 500],
    'class_weight': [{0:0.013, 1:1}, {0:1, 1:1}],
    'max_depth': [5, 10, 15],
    'ccp_alpha': [0.01, 0.025, 0.03]
}
grid_rf = GridSearchCV(RandomForestClassifier(), param_grid=rf_params, verbose=2, n_jobs=-1, scoring='roc_auc')
grid_rf.fit(X_train_transformed, y_train)

# ищем порог, который максимизирует roc-auc score
best_score, best_treshold = 0, np.nan
for t in np.linspace(0.1, 0.9, 20):
    y_pred = grid_rf.predict_proba(X_test_transformed)
    y_pred = y_pred[:, 1] > t
    if roc_auc_score(y_test, y_pred) > best_score:
        best_score, best_treshold = roc_auc_score(y_test, y_pred), t
    
y_pred = grid_rf.predict_proba(X_test_transformed)
y_pred = y_pred[:, 1] > best_treshold
print('Best treshold: ', best_treshold)
print(classification_report(y_test, y_pred))
print('Best roc-auc score: ', roc_auc_score(y_test, y_pred))
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as imbPipeline

# пробуем XGBoost классификатор
weights = [1, 10, 25, 50, 75, 99, 100, 1000]
lrate = [0.1, 0.01, 0.2, 0.3]
max_depth = [5, 10, 20]
alpha = [5, 10, 15, 20]

# применяем KFold кроссвалидацию
param_grid = dict(scale_pos_weight=weights, learning_rate=lrate, max_depth=max_depth, alpha=alpha)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
model = XGBClassifier()
# используем технику SMOTE, то есть oversampling
steps = [('over', SMOTE()), ('model', GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=cv, scoring='roc_auc'))]
pipeline = imbPipeline(steps=steps)
pipeline.fit(X_train_transformed, y_train)
```

```python
# ищем порог, который максимизирует roc-auc score
best_score, best_treshold = 0, np.nan
for t in np.linspace(0.01, 0.9, 50):
    y_pred = pipeline['model'].best_estimator_.predict_proba(X_test_transformed)
    y_pred = y_pred[:, 1] > t
    if roc_auc_score(y_test, y_pred) > best_score:
        best_score, best_treshold = roc_auc_score(y_test, y_pred), t
    
y_pred = pipeline['model'].best_estimator_.predict_proba(X_test_transformed)
y_pred = y_pred[:, 1] > best_treshold
print('Best treshold: ', best_treshold)
print(classification_report(y_test, y_pred))
print('Best roc-auc score: ', roc_auc_score(y_test, y_pred))
```

Таким образом большинство положительных классов определено верно, а так как положительный класс является классом-меньшинством, то нам важно наиболее полно определить именно его, а объекты отрицательного класса, который определились как положительный, не сильно влияют на результат. 
Скорее всего, нужно определить является ли операцияп по счету мошеннической, и тогда нужно заблокировать действительно мошеннические переводы, а случайно заблокированные немошеннические можно разблокировать при дальнейшей свящи с клиентом.
